```python
import pyspark as ps
```

```python
spark = (ps.sql.SparkSession # Create a pyspark session for using spark sql
         .builder            # build it
         .master('local[4]') # using 4 local cpu cores
         .appName('lecture') # named 'lecture'
         .getOrCreate()      # If it already exists, return that object. Else create one.
        )
sc = spark.sparkContext      # Get the spark Context for direct interaction with spark. 
```

```python
tweet_df = spark.read.json('CaseStudies/data/french_tweets.json')
tweet_df.printSchema()
```

```python
tweet_df.take(1)
```

```python
something = tweet_df.select('created_at', 'user.screen_name', 'text', 'retweet_count')
```

```python
type(something)
```

```python
something.show()
```

```python
from pyspark.sql import functions as F
something.agg(F.max("retweet_count")).show()
```

```python
something.select('retweet_count').distinct().count()
```

```python
something.orderBy('retweet_count', ascending = True).show()
```

```python
something.select('screen_name').distinct().show()
```

```python
something.groupBy('screen_name').count().show()
```
